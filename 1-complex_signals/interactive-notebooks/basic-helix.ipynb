```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

We are going to make a complex signal and visualize it in several ways.

```python
# First set simulation parameters
samp_rate = 100 # sampling rate, Hz
fc = 1 # center frequency, Hz
phi = 0 # phase offset, radians
t = 2 # duration, seconds
```

```python
# Make a time array
t_arr = np.arange(0,t,samp_rate**-1) # The array of time samples
```

```python
# Create a complex exponential
basic_exp = np.exp(1j*(2*np.pi*fc*t_arr+phi))
```

In order to explore our signal, we elect to first look at the amplitude and phase of the signal.

```python
fig, ax = plt.subplots(2,1, sharex=True)
ax[0].plot(t_arr, np.abs(basic_exp))
ax[0].set_title('Mag.')
ax[1].plot(t_arr, np.angle(basic_exp))
ax[1].set_title('Phase')
ax[1].set_xlabel('Time (s)')
plt.show()
```

As expected the magnitude is constant and the phase is increasing monotonically. We want to visualize our complex signal in it's entirety though. In order to do this we have to plot it in 3D. 

The 3D plotting functions take x, y, z arrays so we need to decompose the complex signal into those three arrays. We choose to use time for the x-axis, the real part of the signal for the y-axis, and the imaginary part of the signal for the z-axis.

```python
x=t_arr
y=np.real(basic_exp)
z=np.imag(basic_exp)
```

```python
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
ax.plot(x, y, z, linewidth=3, color='blue')
ax.set_title(r"$e^{j(2\pi f_c t + \phi)}$")
ax.set_xlabel(r"$Time (s)$")
ax.set_ylabel(r"$\Im$")
ax.set_zlabel(r"$\Re$")
plt.show()
```

Great! We see our complex signal represented as a helix which is rotating clockwise (if you're facing positive time). We can glean yet more insight by projecting the helix onto the three respective planes.

```python
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
ax.set_title(r"$e^{j(2\pi f_c t + \phi)}$")
ax.set_xlabel(r"$Time (s)$")
ax.set_ylabel(r"$\Im$")
ax.set_zlabel(r"$\Re$")

plot_offset=0.25

ax.plot(x,y,[-1-plot_offset]*len(x), linewidth=1, color='orange')
ax.plot(x,[1+plot_offset]*len(x),z, linewidth=1, color='green')
ax.plot([0-plot_offset]*len(x),y,z, linewidth=1, color='red')

ax.set_xlim([np.min(x)-plot_offset, np.max(x)+plot_offset])
ax.set_ylim([np.min(y)-plot_offset, np.max(y)+plot_offset])
ax.set_zlim([np.min(z)-plot_offset, np.max(z)+plot_offset])

ax.plot(xs=x, ys=y, zs=z, linewidth=3, color='blue')

plt.show()
```

```python
# First set simulation frequencies and other parameters
samp_rate = 100
symbol_rate = 1 
num_periods = 6 # from -3T to 3T
beta=0
```

In the following plot there is a lot going on. I suggest you edit the array of betas so that you're only comparing a single RRC waveform with a single RC waveform. For example, if you use `for beta in [0]:` to only plot the RRC and RC impulse responses for beta=0.

```python
# Impulse response
fig, ax = plt.subplots()
for beta in [0,0.5,1]:
    RC, Ts = IR.RC_freq_domain(symbol_rate, samp_rate, num_periods, beta)
    RRC, _ = IR.RRC_freq_domain(symbol_rate, samp_rate, num_periods, beta)
    plt.plot(Ts,RRC, label=r"RRC $\beta={}$".format(beta))
    plt.plot(Ts,RC, label=r"RC $\beta={}$".format(beta))
plt.legend()
plt.xlabel("Time (T)")
plt.show()
```
